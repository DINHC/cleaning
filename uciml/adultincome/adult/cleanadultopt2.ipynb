```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
```

```python
adultdataopt =  pd.read_csv('data/cleanedadult.csv', sep=",")
adultdataopt.head()
```

```python
adultdataopt2 = adultdataopt.drop(['WorkClass'], axis=1)
adultdataopt.head()
```

```python
adultdataopt2 = adultdataopt2.drop(['NativeCountry'], axis=1)
adultdataopt2.head()
```

```python
adultdataopt2 = adultdataopt2.drop(['Marital-Status'], axis=1)
adultdataopt2.head()
```

```python
adultdataoptdummy = pd.get_dummies(adultdataopt2)
adultdataoptdummy.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = adultdataoptdummy.drop(["Income_ <=50K"], axis='columns').values
y = adultdataoptdummy["Income_ <=50K"].values
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net
number_input_features = len(X_train[0])
hidden_nodes_layer1 =  100
hidden_nodes_layer2 = 50
hidden_nodes_layer3 = 10

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model. I set it initially at an epoch of 100 but eventually realize the accuracy score remain at 100 and loss kept decreasing. 
fit_model = nn.fit(X_train_scaled, y_train, epochs=20)
```

```python
# Evaluating the model using the test data. The result ends up the same as the first attempt. Removing the Work Class, Native Country, and Marital Status 
# produced the same result as opt1. Te
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
