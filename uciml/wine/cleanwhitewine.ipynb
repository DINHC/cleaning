```python
#Creating a test dataframe. The code utilizing np.vstack takes the column names which were stored as a singular column name and moves them down as a row. 0 can now be seen as the column name.
#I chose this method because the code I will use next will split the data into their respective columns using what is available in the row. 
testdf = pd.DataFrame(np.vstack([whitewine.columns, whitewine]))
testdf
```
